# 문자열 다루기
* 문자열은 **sequence**이다.
* Sequence 는 **순번**이 있다.
  * 순번을 **offset index**라고 한다.

  단순하게 순번을 의미한다.
* 파이썬의 인덱스는 두 종류가 있다. (공통점: 정수. 왼쪽에 있는 숫자 제일 작음.) 
  * 양수 인덱스 : 순서
    * 0부터 시작(정반향 인덱스. *왼->오로 이동하며 증가*)
  * 음수 인덱스 : 마지막, 끝에 있는 것 탐색
    * -1부터 역으로 시작(역방향 인덱스. *오->왼으로 이동하며 감소*)
  
  ex. H e l l o 
  
  (양수 인덱스 0 1 2 3 4)

  (음수 인덱스 -5 -4 -3 -2 -1)
  * 인덱스는 무조건 **정수**이다.


```python
text = "Hello World" # 텍스트 문자의 길이는 11 (띄어쓰기 포함)
print(text) 
```

```python
# 시퀀스의 크기(길이)를 잴 수 있는 함수 - len
len(text)
```

```python
# 시퀀스 내의 데이터를 인덱스로 접근하기.  [ ] - braket을 사용
print(text[3])
print(text[6])
```

```python
print(text[10])
```

시퀀스의 제일 마지막 양수 인덱스는 항상 **시퀀스의 크기 -1**

  인덱스는 0부터 시작하기 때문!

  H e l l o 

  길이는 5이지만, 문자열 순서는 0 1 2 3 4이다

```python
print(text[11])   # string index out of range는 인덱스가 범위에서 벗어난 오류. 올바른 인덱스를 넣었는지 확인할 것
```

오류 읽는 습관들이기. 

제일 아래쪽에 오류 내용 확인

문장이 많을 경우, 오류가 난 쪽에 화살표 생성됨

**<오류 읽는 방법>**

1. 제일 아래쪽에 무슨 에러가 났는지 확인
2. 화살표 확인
3. 내가 만들지 않은 코드는 신경 쓰지 않기

```python
print(text[-1]) # -1 인덱스는 언제나 제일 마지막 데이터를 의미한다.
```

```python
print(text[-3])
```

```python
print(text[-11]) # 음수 인덱스는 0이 아닌 -1부터 시작하기 때문에 가능
```

# 슬라이싱(slicing) ★★★★★
데이터를 잘라내고 원하는 데이터를 골라내기 위함
* 인덱스를 사용해서 시퀀스를 **잘라낸다**
* 잘라서 새로운 시퀀스를 만들어 낸다.
  * 단, 시퀀스가 아닌 것은 잘라낼 수 없다.
* [start : end : step] 시작으로부터 마지막 인덱스


```python
text = "안녕하세요!반갑습니다"
print(text)
```

```python
# start -> 0, end -> 3
# 0번 인덱스부터 2번 인덱스까지 추출
text[0 : 3] # 0번 인덱스 ~ 3-1번 인덱스(2번 인덱스)
```

```python
len(text)
```

```python
text[3 : 10]
```

```python
text[3 : 11]
```

* start, end, step은 생략이 가능하다.
* start를 생략하면 자동으로 0이 들어간다. 
* end를 생략하면 자동으로 시퀀스의 길이가 들어간다.

```python
text[3:] # end가 생략되면 문자열의 길이인 11이 자동으로 들어간다. 
# text[3: len(text)]
```

```python
text[:5] # start가 생략되면 첫 번째 인덱스인 0이 자동으로 들어간다.
# text [0:5]
```

```python
text[:-5] # text[0 : -5]
```

step: 지정한 숫자만큼 건너 뛰면서 슬라이싱

step을 생략하면 기본적으로 1이 주어진다.

```python
text = "안녕하세요!반갑습니다"
print(text)
```

```python
text[:] # start : 0, end : len(text), step : 1
```

```python
text[::] # start : 0, end : len(text), step : 1
```

```python
text[::2]
```

```python
text[3 : 8 : 2]
```

```python
text[::-1] # step을 음수로 주면, 역방향으로 추출
```

```python
text[8:3:-1] # step이 음수면 start와 end도 음수로 정해야 함
```
