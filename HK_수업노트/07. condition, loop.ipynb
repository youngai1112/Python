# 제어문
* 코드를 제어하는 문법

##### 조건문 ( Condition )
* 특정한 조건을 만족하면 ( `True` ) 코드를 실행
* `if, elif, else`
  * `if` : 만약에 ~라면
  * `elif` (`else if`) : 그게 아니고 ~라면
  * `else` : 그게 아니면

##### 반복문 ( Loop )
  * 특정 조건을 만족하면 (`True`)코드를 반복
  * `while, for`
    * `while` : 조건제 반복 (`True`면 실행, `False`면 실행 안함)
    * `for` : 순회 반복 (`literable loop`) -Sequence를 순회하면서 반복

### Code Context
* Context : 코드 문맥
* 특정한 상황에 실행해야 할 코드 뭉터기를 의미
* 파이썬에서는 Context를 indent(들여쓰기)로 구분

* indent는 spacebar 2칸이나 4칸으로 설정한다. 
* 처음에 indent를 2칸으로 지정했으면 계속 2칸쓰고, 4칸으로 지정했으면 계속 4칸 쓰면 된다. 
* 개발툴(IDE)에서 자동으로 2칸 4칸 처리 해준다. 


````python
# 조건문, 조건 부분에 bool 타입 데이터 (True / False)가 온다.
# bool 타입이 아닌 데이터가 오면 자동으로 bool 타입으로 바뀐다.
a = 3
b = 2
if a > b :
  print("Hi")
  print("Bye")

print("done")

````

`elif, else`는 `if`에서 조건이 맞지 않았을 때 연속해서 조건검사를 하고 싶을 때 사용한다.

> 만약에 지갑에 12000원 이상 있으면 택시를 타고, 그게 아니고 지갑에 8000 ~ 12000원 미만 있으면 버스를 타고, 그게 아니면 걸어간다.

````python
money = 10000

# if문은 조건문의 시작!
if money >= 10000 : 
  print("택시를 타고 집에 갑니다.")

elif money <= 10000 : 
  print("걸어서 집에 갑니다.")
````

 1. 만약에 돈이 10000원 이상이면 택시 타고, 만약에 돈이 10000원 이하면 걸어간다.

2. 만약에 돈이 10000원 이상이면 택시 타고, 그게 아니고 돈이 10000원 이하면 걸어간다. 


if를 2개 이상 사용해야 하는 경우 (독립조건)
* 만약에 사용자의 계정이 4글자 이상이면 중복검사를 하고, 만약에 비밀번호에 특수문자가 없으면 다시 입력하게 한다. 

`elif`또는 `else`를 사용해야 하는 경우 (연속조건)
  * 만약에 돈이 만원 이상이면 택시 타고, 만원 미만이면 버스탄다. 

* 연속 조건과 독립 조건 구분법
  * 하나의 변수를 여러 번 검사하면 연속 조건인 경우가 많다. 
  * 계정의 길이가 4글자 이상이고, 숫자도 껴있으면 중복버튼 활성화
  * 여러 개의 변수가 연속적으로 이어져야 하는 경우
  * 계정이 조건에 맞고, 비밀번호도 조건에 맞으면 회원가입 버튼을 활성화

## 컴프리헨션
* 입문을 지나서 초보 단계로 넘어가기 위한 관문
* 리스트, set, dict를 간단하게 만들기 위한 문법

반복문을 이용해서 컬렉션에 데이터를 추가하는 경우가 많다!

* 여러 전처리나 효과 없이 아주 단순하게 만들어진 데이터를 컬렉션에 추가(append)라고 싶으면?
* 한줄 짜리 간단한 코드에 의한 아이템 추가
 
 *컴프리헨션은 for문부터 작성하자*


````python
3 ** 2 # 곱하기를 두번하면 제곱 ex. 3 **3은 3의 3제곱
````

````python
# 0~3까지의 각각 값에 제곱한 결과를 리스트에 담아놓기
# 0 - 0, 1 - 1, 2 - 4, 3 - 9

lst =list(range(4)) # [0, 1, 2, 3] 생성
result = [] # 비어있는 리스트 준비, 새로운 컬렉션 생성

for data in lst: # lst에서 무언가를 꺼내올 것이다.
  result.append(data**2)

result
````

기존의 컬렉션을 이용해서 컬렉션 내의 데이터를 이용해 **새로운 컬렉션**을 만들 때 컴프리헨션을 사용한다.

단, 단순한 연산만 있는 경우 사용할 것!

*튜플은 컴프리헨션이 없다*

*컴프리헨션은 for문부터 작성하자*




````python
# 리스트 컴프리헨션
result = [data**2 for data in lst] # data를 lst에서 꺼내오는 부분 -> 맨 앞쪽에 data를 꺼내는 표현/방법/연산, 연산은 한 번만 사용가능, 연산이 두 개 이상일 때 함수 또는 for문 사용
result
````
