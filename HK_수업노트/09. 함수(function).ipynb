매번 ctrl + c, v를 하면서 그때 그때마다 코드를 작성하는게 옳은 방법일까?

````python
# 반복적인 기능을 하는 코드 뭉치(code context)를 정의해 놓고 필요할때마다 사용
# 정의 : def  /// 특정 기능이 있는 코드 뭉치를 만들어 놓겠다는 의미
# 사용(호출) : call /// 만들어 둔 코드 뭉치를 사용하겠다는 의미

# 함수의 정의
def print_grade(score):
  if score >= 90:
    print("A")
  elif score >= 80:
    print("B")
  else:
    print("C")
````

# 함수의 정의 방법


# 함수의 정의 방법
```python
def <fundtion_name>([parameter1, parameter2, ...]):
  ...
  code
  ...
  ```

* 파라미터(Parameter) : 함수 **선언**할 때 호출하는 곳에서 입력해주는 **데이터을 저장할 변수** 
* 아규먼트(Argument) : 함수를 호출하는 곳에서 함수에 넣어주는 데이터 

파라미터의 변수명과 아규먼트의 변수명은 항상 일치해야 할까요?
* No!

````python
# 1. Argument(Parameter) X, Return X

def foo(): # 파라미터가 없다면 ()의 형태로 쓴다
  print("Here is foo")
  print("Parameter X, Return X")
# Parameter가 없으니까 Argument를 안 넣어 준다.
foo() # 아래 결과는 Return이 아니라 print에 의해 출력된 값
````

````python
# 2. Argument(Parameter) O, Return X
def goo(num1, num2): # num1, num2 Parameter가 있음
  if num1 + num2 > 10:
    print("Big")
  else:
    print("Small")

goo(1, 2) # 1, 2는 goo 함수의 Argument
goo(10, 20) # 10, 20은 goo 함수의 Argument
````

````python
# 3. Argument(Parameter) X, Return O
def hoo(): 
  print("Hoo is called")
  lst = range(1, 10)
  total = sum(lst)

  return total # return값이 있음. 함수를 호출한 곳에 값을 줄 것이라는 의미. 함수의 결과물

hoo_result = hoo() # 함수 hoo는 값을 반환한다. 따라서 리턴값을 변수로 받아낼 수 있다. 
                  # hoo()라는 변수의 결과값을 hoo_result의 결과로 집어넣는다.
print("결과 : {}".format(hoo_result))
````

````python
hoo_result
````

````python
# 4. Argument(Parameter) O, Return O
def moo(num1, num2): # Parameter O 
  print("moo is called")
  return num1 + num2 # Return O

a = 10
b = 20
moo_result = moo(a, b) # a는 num1으로, b는 num2로 들어간다.

print(moo_result)
````

# 파라미터를 정의하는 여러 방법
1. `default parameter`
2. 여러 파라미터를 동시에 포함 시키기
  * `*args`
3. 여러 아규먼트를 dict 형태로 받기
  * `**keyword argument`

## Default Parameter ★★★★★
* 함수를 호출할 때 Argument를 넣어주지 않아도 기본적으로 파라미터가 가지고 있어야 할 값을 지정

````python
def noo(a, b, c):
  print(a, b, c)

noo("Hello", 30, 5.1212)
````

함수 `noo`는 Parameter가 3개!
* `noo`를 호출하기 위해서는 Argument를 넣어줘야 한다.

````python
noo(10, 20) # 에러 내용: noo의 Argument는 3개 이다.//위 `noo`의 parameter는 3개
````

````python
# 파라미터 c에 기본적으로 30을 넣어주는 것
# default parameter
def koo(a, b, c=30): 
  print(a, b, c)

koo(10, 20) # c의 기본값이 30으로 지정 되었기 때문에 c에 대한 Argument를 넣지 않아도 koo 호출이 가능
````

````python
koo(10, 20, "Hello")
````

잘못된 default parameter 설정


````python
# 함수를 정의 하면서 오류가 난다
def roo(a, b=20, c): # c가 남음 => 에러 원인
  print(a, b, c,)

roo(10, 30) # 10는 a에, 30은 b에 들어간다. 에러 원인이 아님 
````

default parameter는 항상 오른쪽부터 채워야 한다.

````python
def woo(a, b=20, c=30): # a=□ 여도 무방하다
  print(a, b, c)

# woo 디폴트 파라미터가 2개니까 아규먼트를 1개만 넣어도 호출 된다.
woo(10)
````

## `args`
  * 함수를 정의하면서 Argument가 `N`개가 들어올 수도 있음.

   =>*유동적으로 Argument가 들어오는 상황 발생 가능*
  * `*`을 `asterisk`라고 함


````python
# 2개의 파라미터를 정의
def add_ver1(a,b):
  total = a + b
  print(total)

# 호출할 때 반드시 2개의 아규먼트만을 사용한다.
add_ver1()
````

3개의 아규먼트를 받아서 더하고 싶으면?

````python
def add_ver2(a, b, c):
  total = a + b + c

  return total

add_result = add_ver2(10, 20, 30)
print(add_result)
````

`N`개의 Argument를 받아서 더하고 싶으면? `*args`

````python
def add_ver3(*args):
  # print(type(args), args)
  # print(args[3])
  total = sum(args)
  print(total)

add_ver3(10, 20, 30, 40, 50, 60, 70, 80) # 여러 개의 아규먼트를 동시에 받아내게 되면, 동시에 튜플로 패킹해서 받아들이게 된다. 하나로 묶여서 *args라는 변수로 들어가게 된다.
````

````python
print("Hi", "Hello", "Bye") # print라는 함수도 아규먼트로 구현되어 있다.
````

하나의 함수에서 여러 개의 `*args`를 사용할 수 있을까?

````python
def goo(*args, *args2): # 어디서부터 어디까지 args1인지, args2인지 구분을 못하게 된다.
  print(args1)
  print(args2)
````

`positional argument`와 `*args` 같이 사용하기

*Positional Argument는 Parameter의 위치에 맞게 들어가는 Argument

````python
def qoo(a, b, *args): # `*`은 선언할때만 사용 /// 출력에는 사용하지 않는다.
  print(a)
  print(b)
  print(args)

qoo(1,2,3,4,5,6,7,8,9) # 1는 a에, 2는 b에, 나머지 3~9까지는 args로 들어간다.
````

````python
# *args가 가운데에 위치할 수 있을까요?
def eoo(a, *args, b):
  print(a)
  print(b)
  print(args)
````

````python
eoo(1,2,3,4,5,6,7,8,9) # 에러 내용: 1은 a에 들어가지만, 나머지는 2~9까지로 판단, b에 들어갈 값이 없다.
````

````python
# b는 Key Word Argument로 설정
eoo(1,2,3,4,5,6,7,8,9, b="Hi") # b는 끝부분에 위치해야 한다.
                               # 1은 a에, 2~9까지 c에, "Hi"는 b에 
````

### Keyword Argument?

파라미터에 값을 넣어줄 때 파라미터의 변수명을 같이 기입하는 것

````python
def too(a, b, c):
  print(a, b, c)

too(10, 20, 30) # 파라미터 a,b,c의 순서에 맞춰서 아규먼트를 집어 넣은 것(Positional Argument)
too(b=20, c=30, a=10) # 파라미터의 이름을 기입해서 집어 넣기
````

## `**kwargs`
* 입력된 keyword argument들을 dict로 정리

````python
def yoo(a, b, **kwargs):
  print(a)
  print(b)
  print(kwargs)
````

````python
yoo(10, 20, c=30, d=40, e=50)
````

keyword argument의 변수명이 `key`로, r값이 `value`로 `**kwargs`에 모아진다.

````python
def aoo(a, b, *args, **kwargs): # 1은 a에, 2는 b에, 3~8까지 args에, kwargs는 명시된 것이 없어 X
  print(a)
  print(b)
  print(args)
  print(kwargs)

aoo(1,2,3,4,5,6,7,8)
print()
aoo(1,2,3,4,5,6,7, c=30, e=40)
````

````python
# keyword argument도 항상 오른쪽부터 채운다
aoo(1, 2, 3, a=10, 5, 6, 7)
````

````python
# default parameter를 사용할 이유가 없다.
def boo(a, b, c=30, d=40, *args, **kwargs): # 이렇게 사용하는 경우는 거의 없다
  print(a)
  print(b)
  print(c)
  print(d)
  print(args)
  print(kwargs)
````

````python
boo(1, 2, 3, 4, 5, 6, 7, e=10, f=30)
````
