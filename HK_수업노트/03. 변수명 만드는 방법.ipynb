# 변수 네이밍 규칙
  * 관례
  * 단어와 단어가 이어지는 사이를 _로 이어주는 snake case를 활용


```python
# 1. 변수와 제일 앞글자에 특수 문자는 올 수 없다.(_ 제외)
_student = 'MINHO' # 상수처럼 변수를 사용해야 할 때 사용
```

```python
^a = 10 # invalid syntax : 문법 오류(잘못된 글자를 적음) 
```

```python
# 2. 숫자도 변수명 제일 앞에 올 수 없다.
^a = 3
```

**invalid syntax** 오류는 보통 오타 때문에 난다

# 사용하면 안되는 변수의 이름
* reserved kerwords (예약어)
  * 파이썬에서 미리 선점하여 사용중인 키워드
  * 변수, 함수, 클래스 등등의 **사용자 정의 이름**으로 사용이 불가
 
* 드라이브 자료 참고 - 체득되는 내용이므로, 암기 노노
