# 객체 (Object)
클래스를 기반으로 해서 실제 사용할 수 있게된 **것 (실물, 실체가 있는)**



````python
# Car 클래스를 정의
class Car:

  # 멤버 변수 정의
  def __init__(self):
    self.car_color = "Black"
    self.car_brand = "Audi"
    self.car_price = 9000

  # 메소드 정의
  def drive(self):
    print("차가 앞으로 갑니다.")

  def stop(self):
    print("차가 멈췄습니다.")
````

````python
# Car 클래스를 기반으로 Car 객체 만들기
my_car = Car()
````

````python
my_car.drive() # 앞으로 가는 기능을 수행할 수 있게 됨. 실제로 사용할 수 있도록 객체를 만듦
my_car.stop()
````

````python
Car.drive() # 사용할 수 없다. 개념만 있으므로
````

# 클래스에 Parameter 정의하기

````python
class Car2:

  # method에 parameter 정의
  def __init__ (self, car_color, car_brand, car_price):
    self.car_color = car_color
    self.car_brand = car_brand
    self.car_price = car_price

    # parameter 이름과 멤버 변수 이름을 굳이 맞출 필요는 없으나, 어지간하면 맞춰라
    # self 가 앞에 붙어 있으면 멤버 변수
    # self 가 앞에 붙어 있지 않으면 파라미터, 지역변수

  def drive(self, where): # where는 멤버변수가 아닌 parameter이고, 객체를 이루고 있는 데이터가 아니다. 특정 행동 또는 기능을 수행할 때만 필요하다.
    print("자동차가 앞으로 갑니다.")
    print(f"목적지는 {where}입니다.")
    print("제 자동차는 {}에서 만든 {}색 {}짜리 자동차입니다.".format(self.car_brand, self.car_color, self.car_price))
````

````python
my_car2 = Car2(car_color="파란", car_brand="현대", car_price=5000)
````

````python
my_car2.drive("양평(경기도양평아님)")
````

````python
my_car2.drive()
````

````python
my_car3 = Car2("초록", "기아", 3000)
my_car3.drive("우리집")
````

# 멤버
* 클래스가 객체가 되었을 때 사용할 수 있으면 **멤버**
* 멤버들(멤버 변수, 멤버 메소드)는 `self`를 이용해서 접근이 가능
* 멤버들은 클래스 내부 어디에서든 `self`로 접근이 가능

## `self`
* 객체 자기 자신 

````python
my_car2.car_brand # 이 차(my_car2)의 브랜드는 현대에요
````

````python
my_car2.car_price # `.`은 접근 연산자(~의). my_car2의 car_price라는 의미
````
